````python
# Update the MYSQL Connection Details in the python script
import requests
from bs4 import BeautifulSoup
import re
import json
import concurrent.futures    #  I'll enhance the script to handle a list of 100 websites and ensure robust error handling. That's why I use of 'concurret.futures'
                            # to speed up the process by running multiple requests in parallel.

# Function to extract meta information
def extract_meta(soup):
    title = soup.find('title').text if soup.find('title') else "N/A"
    description = soup.find('meta', attrs={'name': 'description'})
    description = description['content'] if description else "N/A"
    return title, description

# Function to extract social media links
def extract_social_media(soup):
    social_media_domains = ["facebook.com", "twitter.com", "linkedin.com", "instagram.com", "youtube.com"]
    links = [a['href'] for a in soup.find_all('a', href=True)]
    social_links = [link for link in links if any(domain in link for domain in social_media_domains)]
    return social_links

# Function to extract payment gateways
def extract_payment_gateways(html):
    payment_keywords = ["paypal", "stripe", "razorpay"]
    return [keyword for keyword in payment_keywords if keyword in html.lower()]

# Function to extract tech stack
def extract_tech_stack(soup):
    tech_keywords = {
        'js_frameworks': ["angular", "react", "vue", "jquery"],
        'cms': ["wordpress", "joomla", "drupal"],
        'mvc': ["django", "ruby on rails", "laravel"]
    }
    scripts = [script.get('src', '') for script in soup.find_all('script')]
    links = [link.get('href', '') for link in soup.find_all('link')]
    combined = " ".join(scripts + links)
    detected_tech = {key: [tech for tech in techs if tech in combined.lower()] for key, techs in tech_keywords.items()}
    return detected_tech

# Function to determine website language
def extract_language(soup):
    html_tag = soup.find('html')
    return html_tag.get('lang') if html_tag and html_tag.has_attr('lang') else 'N/A'

# Function to extract website category (simplified example based on keywords)
def extract_category(soup):
    categories = {
        'ecommerce': ["cart", "checkout", "product"],
        'blog': ["blog", "post", "article"],
        'news': ["news", "report", "journal"]
    }
    text = soup.get_text().lower()
    detected_categories = [category for category, keywords in categories.items() if any(keyword in text for keyword in keywords)]
    return detected_categories if detected_categories else ["N/A"]

# Main function to scrape website
def scrape_website(url):
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()  # Raise an error for bad status codes
        soup = BeautifulSoup(response.content, 'lxml')

        title, description = extract_meta(soup)
        social_links = extract_social_media(soup)
        payment_gateways = extract_payment_gateways(response.text)
        tech_stack = extract_tech_stack(soup)
        language = extract_language(soup)
        category = extract_category(soup)

        return {
            'url': url,
            'title': title,
            'description': description,
            'social_links': social_links,
            'payment_gateways': payment_gateways,
            'tech_stack': tech_stack,
            'language': language,
            'category': category
        }
    except requests.RequestException as e:
        return {'url': url, 'error': str(e)}

# List of 100 websites
urls = [
    # Add 100 websites here
    "https://www.kaggle.com/",
    "https://data.gov/",
    "https://jsonplaceholder.typicode.com/guide/",
    "https://github.com/geertjanw/ojet-training/blob/master/employeeData.json",
    "https://www.google.com/",
    "https://catalog.data.gov/dataset/?res_format=JSON",
    "https://www.google.com/search?q=paystore&oq=paystore&gs_lcrp=EgZjaHJvbWUyBggAEEUYOTIPCAEQABgKGIMBGLEDGIAEMg8IAhAAGAoYgwEYsQMYgAQyDwgDEAAYChiDARixAxiABDIJCAQQABgKGIAEMgkIBRAAGAoYgAQyDwgGEAAYChiDARixAxiABDIGCAcQBRhA0gEJMTEyMThqMGo3qAIIsAIB&sourceid=chrome&ie=UTF-8",
    "https://www.youtube.com/",
    "https://www.whatsapp.com",
    "https://www.tcs.com/careers/india/tcs-fresher-hiring-nqt-2024",
    "https://www.browserstack.com/guide/selenium-framework#:~:text=Selenium%20Framework%20is%20a%20suite,lots%20of%20time%20and%20effort.",
    "https://www.jai-kisan.com/",
    # ... (90 more websites)
]

# Run scraping in parallel
results = []
with concurrent.futures.ThreadPoolExecutor(max_workers=10) as executor:
    futures = [executor.submit(scrape_website, url) for url in urls]
    for future in concurrent.futures.as_completed(futures):
        results.append(future.result())

# Output the results to a JSON file
with open('website_data.json', 'w') as f:
    json.dump(results, f, indent=4)

# Print the results
print(json.dumps(results, indent=4))
````

# Web Scraper for Website Information Extraction

## Introduction

This project extracts specific information from a list of websites and stores it in a MySQL database.

## Prerequisites

- Python 3.x
- MySQL Server
- Python libraries: `beautifulsoup4`, `selenium`, `pymysql`, `sqlalchemy`, `requests`
- ChromeDriver (for Selenium)

## Setup

1. **Install Dependencies**

    ```bash
    pip install beautifulsoup4 selenium pymysql sqlalchemy requests
    ```

2. **Setup MySQL Database**

    - Create a database and table using the provided SQL script `create_tables.sql`.

    ```sql
    CREATE DATABASE IF NOT EXISTS website_data;

    USE website_data;

    CREATE TABLE IF NOT EXISTS website_info (
        id INT AUTO_INCREMENT PRIMARY KEY,
        url VARCHAR(255) UNIQUE,
        social_media_links TEXT,
        tech_stack TEXT,
        meta_title VARCHAR(255),
        meta_description TEXT,
        payment_gateways TEXT,
        language VARCHAR(255),
        category VARCHAR(255)
    );
    ```

3. **Configure Database Connection**

    - Update the `DATABASE_URI` in the Python script to match your MySQL credentials.

    ```python
    DATABASE_URI = 'mysql+pymysql://username:password@localhost:3306/website_data'
    ```

4. **Download ChromeDriver**

    - Download ChromeDriver from [here](https://sites.google.com/a/chromium.org/chromedriver/downloads) and place it in your system path or specify the path in the script.

5. **Run the Script**

    ```bash
    python scrape_websites.py
    ```

## Approach

1. **Web Scraping**: We used BeautifulSoup for parsing HTML and Selenium for handling JavaScript-heavy websites.
2. **Data Extraction**: Extracted meta information, social media links, tech stack, payment gateways, language, and category.
3. **Data Storage**: Stored the extracted data in a MySQL database using SQLAlchemy.

## Challenges

1. **Dynamic Content**: Handling JavaScript-heavy websites required Selenium, which increased the complexity and runtime.
2. **Data Consistency**: Ensuring consistency and handling edge cases, such as missing meta tags or non-standard HTML structures.

## Conclusion

This project provides a robust framework for extracting and storing structured information from websites. Further improvements can include better handling of dynamic content and more sophisticated category extraction.
